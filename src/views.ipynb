```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def plot_predictions(csv_file):
    # 读取 CSV 文件
    df = pd.read_csv(csv_file)

    # 提取文件名（不含扩展名）
    filename = os.path.splitext(os.path.basename(csv_file))[0]

    # 绘制折线图
    plt.figure(figsize=(10, 6))
    for column in df.columns:
        plt.plot(range(1, len(df) + 1), df[column], label=column)
    
    plt.xlabel('Case')
    plt.ylabel('Prediction Probability')
    plt.title('Predictions')
    plt.legend()
    plt.grid(True)
    
    # 添加文件名注释
    plt.annotate(filename, xy=(0.5, -0.15), xycoords="axes fraction", ha="center", va="center")
    
    plt.show()

```

### AlexNet

```python
# 调用函数来绘制折线图

plot_predictions('C:/Users/30961/Desktop/test/newtest/MRNet/models/AlexNet/losses_axial.csv')

plot_predictions('C:/Users/30961/Desktop/test/newtest/MRNet/models/AlexNet/losses_coronal.csv')

plot_predictions('C:/Users/30961/Desktop/test/newtest/MRNet/models/AlexNet/losses_sagittal.csv')
```
